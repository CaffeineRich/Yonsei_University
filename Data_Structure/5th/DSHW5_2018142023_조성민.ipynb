#### PROBLEM I-1: TriNode (30 Points)

 Implement $\texttt{TriNode}$.  
> In previous assignments, we exploited $\texttt{dNode}$ in order to implement $\texttt{DoublyLinkedList}$.  
> Likewise, we will exploit $\texttt{TriNode}$ in order to implement $\texttt{BinaryHeap}$.  
> 
> $\texttt{TriNode}$ contains an item and **<U>three</U>** references to the next node.  

> Below are specifications of $\texttt{TriNode}$ operations.
>> <code>\_\_init\_\_(data)</code>: constructs a TriNode with given <code>data</code>, and default references set to <code>None</code>.  
>> <code>set_refer(loc, other_TriNode)</code>: sets reference corresponding to given <code>loc</code>, where <code>loc</code> $\in$ $\{$ 'p'(=parent), 'r'(=right), 'l'(=left) $ \} $.  
>> <code>get_data()</code>: returns data stored in the node.

### Answer:

```python
class TriNode:
    def __init__(self, data):
        self.data=data
        self.parent=None
        self.left=None
        self.right=None
        
    def set_refer(self, loc, other_TriNode):
        if loc=='p':
            self.parent=other_TriNode
        if loc=='l':
            self.left=other_TriNode
        if loc=='r':
            self.right=other_TriNode
            
    def get_data(self):
        return self.data
```

```python
class TreeNode:
    def __init__(self, key, val, left = None, right = None, parent = None):
        self.key = key
        self.payload = val
        self.left_child = left
        self.right_child = right
        self.parent = parent
    def is_left_child(self):
        return self.parent and self.parent.left_child == self    
    def is_right_child(self):
        return self.parent and self.parent.right_child == self    
    def is_root(self):
        return not self.parent    
    def is_leaf(self):
        return self.right_child is None and self.left_child is None    
    def has_any_children(self):
        return self.right_child is not None or self.left_child is not None  
    def has_both_children(self):
        return self.right_child is not None and self.left_child is not None
    def replace_node_data(self, key, value, lc, rc): 
        self.key = key 
        self.payload = value 
        self.left_child = lc 
        self.right_child = rc
        if self.left_child is not None:
            self.left_child.parent = self 
        if self.right_child is not None:
            self.right_child.parent = self            
    def __iter__(self):
        if self:
            if self.left_child is not None:
                for elem in self.left_child:
                    yield elem
            yield self.key
            if self.right_child is not None:
                for elem in self.right_child:
                    yield elem
            
class BinarySearchTree: 
    def __init__(self):
        self.root = None
        self.size = 0        
    def __len__(self): 
        return self.size    
    def __iter__(self):
        return self.root.__iter__()    
    
    def put(self,key,val):
        if self.root:
            self._put(key,val,self.root)
        else:
            self.root=TreeNode(key,val)
        self.size=self.size+1
    def _put(self,key,val,current_node):
        if key<current_node.key:
            if current_node.left_child:
                self._put(key,val,current_node.left_child)
            else:
                current_node.left_child=TreeNode(key,val,parent=current_node)
        else:
            if current_node.right_child:
                self._put(key,val,current_node.right_child)
            else:
                current_node.right_child=TreeNode(key,val,parent=current_node)
    def __setitem__(self,k,v):
        self.put(k,v)
    def get(self,key):
        if self.root:
            res=self._get(key,self.root)
            if res:
                return res.payload
            else:
                return None
        else:
            return None
    def _get(self,key,current_node):
        if not current_node:
            return None
        elif current_node.key==key:
            return current_node
        elif key<current_node.key:
            return self._get(key,current_node.left_child)
        else:
            return self._get(key,current_node.right_child)
    def __getitem__(self,key):
        return self.get(key)
    def __contains__(self,key):
        if self._get(key,self.root):
            return True
        else:
            return False
    def delete(self,key):
        if self.size>1:
            node_to_remove=self._get(key,self.root)
            if node_to_remove:
                self.remove(node_to_remove)
                self.size=self.size-1
            else:
                raise KeyError('Error, key not in tree')
        elif self.size==1 and self.root.key==key:
            self.root=None
            self.size=self.size-1
        else:
            raise KeyError('Error, key not in tree')
            
    def __delitem__(self,key):
        self.delete(key)
        
    def remove(self, current_node):      
        if current_node.is_leaf():
            if current_node == current_node.parent.left_child:
                current_node.parent.left_child = None
            else:
                current_node.parent.right_child = None
                
        elif current_node.has_both_children():
            succ = current_node.find_successor() 
            succ.splice_out()
            current_node.key = succ.key 
            current_node.payload = succ.payload    
            
        else:
            if current_node.has_left_child():
                if current_node.is_left_child(): 
                    current_node.left_child.parent = current_node.parent 
                    current_node.parent.left_child = current_node.left_child 
                    
                elif current_node.is_right_child():
                    current_node.left_child.parent = current_node.parent
                    current_node.parent.right_child = current_node.left_child
                    
                else: 
                    current_node.replace_node_data(
                      current_node.left_child.key,
                      current_node.left_child.payload,
                      current_node.left_child.left_child,
                      current_node.left_child.right_child)       
                    
            else:
                if current_node.is_left_child():
                    current_node.right_child.parent = current_node.parent
                    current_node.parent.left_child = current_node.right_child
                    
                elif current_node.is_right_child(): 
                    current_node.right_child.parent = current_node.parent
                    current_node.parent.right_child = current_node.right_child
                    
                else: 
                    current_node.replace_node_data(
                      current_node.right_child.key,
                      current_node.right_child.payload,
                      current_node.right_child.left_child,
                      current_node.right_child.right_child)
```

```python
def ubst2bbst(unbalanced_tree):
    balanced_tree=BinarySearchTree()
    a_list=[]
    key=0
    tree_len=len(unbalanced_tree)
    while tree_len>0:
        if unbalanced_tree[key]!=None:
            tree_len-=1
            a_list.append(key)
        key+=1
        
    arr_key=arrange_keys(a_list)
    d={}
    for i in arr_key:
        d[i]=unbalanced_tree[i]
     
    for k,v in d.items():
        balanced_tree[k]=v
    
    return balanced_tree
    
    
# Define your addtional functions if necessary
def arrange_keys(a_list):
    b_list=[]
    length=len(a_list)
    while length>0:
        b_list.append(a_list[length//2])
        a_list[length//2]=None
        length=length//2
        
    i=0
    while i<len(a_list)//2:
        if a_list[i]==None:
            i+=1
        else:
            b_list.append(a_list[i])
            a_list[i]=None
            i+=1
            
    start_value=len(a_list)//2
    while (start_value+len(a_list))//2 < len(a_list)-1:
        b_list.append(a_list[(start_value+len(a_list))//2])
        a_list[(start_value+len(a_list))//2]=None
        start_value=(start_value+len(a_list))//2
    b_list.append(a_list[len(a_list)-1])
    a_list[len(a_list)-1]=None
    
    j=len(a_list)//2
    while j<len(a_list):
        if a_list[j]==None:
            j+=1
        else:
            b_list.append(a_list[j])
            a_list[j]=None
            j+=1
    return b_list

```

**Problem Description**
> $ DFS : \texttt{Depth-first Search}$
>
> It starts at the root node and traverses as far as possible along each branch. If it arrives at a leaf node, it backtracks and moves on to unvisited branch.
>
> When it encounters a target item, it returns the list that consists of visited nodes(key).
>
> Otherwise, raise Error

> $ BFS  : \texttt{Breadth-first Search}$
>
> It starts at the root node and traverses all of the neighbor nodes at the present depth.  
> If there is no taget item at the present depth, it moves on to the nodes at the next depth level. 
> 
> When it encounters a target item, it returns the list that consists of visited nodes(key)
>
> Otherwise, raise Error

> For further study, refer to [DFS](https://en.wikipedia.org/wiki/Depth-first_search) and [BFS](https://en.wikipedia.org/wiki/Breadth-first_search)

```python
def DFS(tree, target_item):
    tree=tree.root
    visited=[]
    found=False
    while not found and tree:
        if tree.payload==target_item:
            found=True
        else:
            if tree.is_leaf():
                visited.append(tree.key)
                tree=tree.parent
            elif tree.has_both_children():
                if tree.left_child.key not in visited and tree.right_child.key not in visited:
                    visited.append(tree.key)
                    tree=tree.left_child
                elif tree.left_child.key in visited and tree.right_child.key not in visited:
                    tree=tree.right_child
                elif tree.left_child.key in visited and tree.right_child.key in visited:
                    tree=tree.parent
            elif tree.left_child and not tree.right_child:
                if tree.left_child.key not in visited:
                    visited.append(tree.key)
                    tree=tree.left_child
                else:
                    tree=tree.parent
            
            elif tree.right_child and not tree.left_child:
                if tree.right_child.key not in visited:
                    visited.append(tree.key)
                    tree=tree.right_child
                else:
                    tree=tree.parent
    
    if found:
        return visited

    raise ValueError('There is no such item')
    
def BFS(tree, target_item):
    tree=tree.root
    queue=[tree.key]
    visited=[]
    found=False
    done=False
    done1=False
    case=False
    while not found and queue:
        if tree.payload==target_item:
            if not case:
                visited.append(queue.pop(0))
            found=True
        else:
            if tree.is_leaf():
                visited.append(queue.pop(0))
                done=True
                
                while tree.parent:
                    tree=tree.parent
                
            elif tree.has_both_children():
                if tree.left_child.key not in visited and tree.right_child.key not in visited and tree.left_child.key not in queue and tree.right_child.key not in queue:
                    queue.append(tree.left_child.key)
                    queue.append(tree.right_child.key)
                    
                    if tree.key not in visited and tree.key in queue:
                        if tree.parent==None:
                            tree=tree.left_child
                            visited.append(queue.pop(0))
                        else:
                            tree=tree.parent
                            visited.append(queue.pop(0))
                    
                    elif tree.key in visited and tree.key not in queue:
                        if tree.left_child.key not in visited:
                            tree=tree.left_child
                            
                elif tree.left_child.key not in visited and tree.right_child.key not in visited and tree.left_child.key in queue and tree.right_child.key in queue:      
                    if tree.key in visited and tree.key not in queue:
                        tree=tree.left_child
                    
                elif tree.left_child.key in visited and tree.right_child.key not in visited and tree.right_child.key in queue:
                    tree=tree.right_child
                    
                    
                elif tree.left_child.key in visited and tree.left_child.key not in queue and tree.right_child.key in visited and tree.right_child.key not in queue:
                    if not done:
                        tree=tree.left_child
                    else:
                        tree=tree.right_child
                    
                    
                    
                    
            elif tree.left_child and not tree.right_child:
                if tree.left_child.key not in visited and tree.left_child.key not in queue:
                    queue.append(tree.left_child.key)
                if tree.left_child.key not in visited and tree.left_child.key in queue:
                    tree=tree.parent
                    visited.append(queue.pop(0))
                    
            elif tree.right_child and not tree.left_child:
                if tree.is_root():
                    case=True
                if tree.right_child.key not in visited and tree.right_child.key not in queue:
                    queue.append(tree.right_child.key)
                if tree.right_child.key not in visited and tree.right_child.key in queue:
                    if not case:
                        if not done1:

                            if not done1:
                                done1=True
                            else:
                                done1=False
                            if tree.parent:
                                tree=tree.parent
                                
                            visited.append(queue.pop(0))

                        else:
                            tree=tree.right_child
                    else:
                        tree=tree.right_child
                        visited.append(queue.pop(0))
                        
                        
                if case and tree.right_child.key in visited:
                    tree=tree.right_child
                   
    if found:
        return visited

    raise ValueError('There is no such item')
```
